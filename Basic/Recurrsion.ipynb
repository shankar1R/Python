# Recurrsion in python

```python
# use the function only one time print ("hello")
def greet():
    print("hello")
greet()
```

```python
import sys

print(sys.getrecursionlimit()) # limit is 3000 only 

def greet():
    print("hello")
    greet()
```

```python
import sys
sys.setrecursionlimit(2000)

print(sys.getrecursionlimit()) # limit is 3000 only 

i=0
def greet():
    print("hello",i)
    greet()
greet()
```

# Factorial using Recurrsion

```python
def fact(n):

    if n==0:
        return 1
    
    return n*fact(n-1)

result=fact(10)
print(result)
```
